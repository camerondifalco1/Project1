```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\CDiFa\Project1\Resources\resources - schooling edit.csv'
df = pd.read_csv(file_path)

# Sort the DataFrame by country names (A-Z)
df_sorted = df.sort_values('Country')

# Calculate the average school per country
average_school = df.groupby('Country')['Schooling'].mean()

# Create the plot
fig, ax = plt.subplots()
ax.plot(df_sorted['Country'], df_sorted['Schooling'], marker='o', linestyle='', label='Schooling Level')

# Set axis labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Schooling Level')
ax.set_title('Schooling Level by Country')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add legend
ax.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\CDiFa\Project1\Resources\resources - schooling edit.csv'
df = pd.read_csv(file_path)

# Sort the DataFrame by country names (A-Z)
df_sorted = df.sort_values('Country')

# Set colors based on status
colors = {'Developed': 'blue', 'Developing': 'red'}
df_sorted['Color'] = df_sorted['Status'].map(colors)

# Calculate the average school per country
average_school = df.groupby('Country')['Schooling'].mean()

# Create the plot
fig, ax = plt.subplots()
ax.scatter(range(len(df_sorted)), df_sorted['Schooling'], c=df_sorted['Color'], marker='o', label='Developed Vs         Developing')

# Set axis labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Schooling Level')
ax.set_title('Schooling Level by Country')

# Remove country names from x-axis
ax.set_xticks([])
ax.set_xticklabels([])

# Add legend
ax.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\CDiFa\Project1\Resources\resources - schooling edit.csv'
df = pd.read_csv(file_path)

# Create the plot
fig, ax = plt.subplots()
ax.scatter(df['happiness_score'], df['Schooling'], marker='o', c='blue', label='Scatter Plot')

# Set axis labels and title
ax.set_xlabel('Happiness Score')
ax.set_ylabel('Schooling Level')
ax.set_title('Schooling Level vs. Happiness Score')

# Add legend
ax.legend()

# Display the plot
plt.show()
```
